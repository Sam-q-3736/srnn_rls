```python
from run_QIF import *
from train_QIF import *
from genw_sparse import *
```

```python
def gen_rand_stim(N, timesteps):
    stim = np.zeros((N, timesteps))
    for row in range(N):
        rstim = 2 * sp.stats.uniform.rvs(0, 1) - 1 # random stim weight from -1, 1
        stim[row][int(stim_on/dt):int(stim_off/dt)] = rstim
    return stim
```

```python
def gen_sine_targets(N, timesteps, max_cycles): 
    targets = np.zeros((N, timesteps))
    for i in range(N):
        start = sp.stats.uniform.rvs() * 2 * np.pi * max_cycles
        stop = sp.stats.uniform.rvs() * 2 * np.pi * max_cycles
        x = np.linspace(start, stop, timesteps)
        targets[i] = np.sin(x)
    return targets
```

# training

```python
# parameters
N = 200
m = 0
std = 4
p = 0.3

T = 1000 # ms 
stim_on = 0 # ms
stim_off = 50 # ms
dt = 0.1 # ms
nloop = 10
train_every = 2 # ms

timesteps = int(T/dt) # unitless, total number of timesteps

tau = 1 # ms, neuron phase decay constant
tau_s = 20 # ms, spike train decay constant 
lam = 1 # learning rate constant

neuron_params = N, tau, tau_s, lam
time_params = T, stim_on, stim_off, dt
train_params = nloop, train_every
```

```python
W = genw_sparse(N, m, std/np.sqrt(N*p), p)
W_trained = np.copy(W)
stim = gen_rand_stim(N, timesteps)
targets = gen_sine_targets(N, timesteps, 2) # 2 is maximum number of cycles in training period
```

```python
W_trained, spks, sdrive, thetas, spk_rast = train_QIF(neuron_params, time_params, train_params, W_trained, stim, targets)
```

```python
ind = 4

fig, axs = plt.subplots(nloop, 1, figsize = (3, 12), sharex = True)
for i in range(nloop):
    axs[i].plot(sdrive[ind][i*int(T/dt):(i+1)*int(T/dt)])
    axs[i].plot(targets[ind])
    axs[i].plot(stim[ind])
axs[0].set_title("Synaptic drive over training")
```

```python
corrs = np.zeros(nloop)
for t in range(nloop):
    corr_sum = np.zeros((2, 2))
    for i in range(N):
        corr_sum += np.corrcoef(sdrive[i][t*int(T/dt):(t+1)*int(T/dt)], targets[i]) 
    
    corrs[t] = corr_sum[0][1] / N

plt.plot(corrs)
plt.ylim(0, 1)
plt.title("Correlation across training trials")
plt.ylabel("Correlation")
plt.xlabel("Training trial")
```

```python
final_corrs = np.zeros(N)
for i in range(N):
    final_corrs[i] = np.corrcoef(sdrive[i][(nloop - 1)*int(T/dt):nloop*int(T/dt)], targets[i])[0, 1]

plt.hist(final_corrs, bins = np.linspace(0, 1, 50));
plt.title("Correlation histogram of units and targets")
plt.xlabel("Correlation")
plt.ylabel("Number of units")
```
