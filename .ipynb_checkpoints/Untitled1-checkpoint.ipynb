```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#parameters
N = 1
m = 0
std = 1
p = 0.5

T = 100 # ms 
stim_on = 0 # ms
stim_off = 100 # ms
dt = 0.01 # ms
timesteps = int(T/dt) # unitless, total number of timesteps
tau = 1 # ms, neuron phase decay constant
tau_s = 20 # ms, spike train decay constant 
```

```python
stim = np.ones((1, timesteps))
```

```python
# initial dynamics
W = [1]
theta = np.zeros(N) # phase of neurons
u = np.zeros(N) # synaptic drive
r = np.zeros(N) # filtered synaptic drive
```

```python
# ODE of theta neuron model
def dtheta(theta, n_drive):
    return 1/tau * (1 - np.cos(theta) + np.multiply(n_drive, (1 + np.cos(theta))))

# ODE of filtered spike train
def dr(r): 
    return -1/tau_s * r # does not include the addition of new spikes
```

```python
# optional variable to track states
spks = []
sdrive = []
thetas = []

# training loop
t = 0.0
while t < T:

    # RK4 for theta neuron model
    k1 = dt * dtheta(theta, u + stim[:, int(t/dt)]);
    k2 = dt * dtheta(theta + k1/2, u + stim[:, int(t/dt)]);
    k3 = dt * dtheta(theta + k2/2, u + stim[:, int(t/dt)]);
    k4 = dt * dtheta(theta + k3, u + stim[:, int(t/dt)]);
    theta_next = theta + (k1 + 2*k2 + 2*k3 + k4)/6;
    
    # RK4 for filtered spike train 
    r1 = dt * dr(r) 
    r2 = dt * dr(r + r1/2)
    r3 = dt * dr(r + r2/2)
    r4 = dt * dr(r + r3)
    r_next = r + (r1 + 2*r2 + 2*r3 + r4)/6
    
    # spike detection
    idx1 = theta_next - theta > 0
    idx2 = theta_next - theta > np.mod(np.pi - np.mod(theta, 2*np.pi), 2*np.pi) # surely there's a better way to do this...
    idx = np.multiply(idx1, idx2)
    r[idx] += 1/tau_s # update spikes, what are units? should be millivolts
    
    # update variables
    theta = theta + (k1 + 2*k2 + 2*k3 + k4)/6;
    r = r + (r1 + 2*r2 + 2*r3 + r4)/6
    u = np.dot(W, r)
    t = t + dt
    
    # track variables (optional)
    spks.append(r)
    sdrive.append(u)  
    thetas.append(theta)
  
    # train W matrix
    # only train after initial stimulus
    if t > stim_off: 
        # update each row of W by RLS
        for row in range(N):
            ;

thetas = np.transpose(thetas)
spks = np.transpose(spks)
sdrive = np.transpose(sdrive)
```
